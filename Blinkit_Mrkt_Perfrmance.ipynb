```python
import duckdb
```

```python
duckdb.sql("SELECT * FROM read_csv('blinkit_marketing_performance.csv')")
```

```python
duckdb.sql("CREATE TABLE mrkt_performance as SELECT * FROM read_csv('blinkit_marketing_performance.csv')")
```

```python
duckdb.sql("SELECT * FROM mrkt_performance")
```

```python
duckdb.sql("SELECT DISTINCT target_audience FROM mrkt_performance")
```

```python
duckdb.sql("SELECT DISTINCT campaign_name FROM  mrkt_performance")
```

**Understand Marketing Reach:**

```python
duckdb.sql("SELECT DISTINCT channel,SUM(impressions) AS Total_Reach FROM mrkt_performance GROUP BY channel")
```

```python
duckdb.sql("SELECT DISTINCT channel, MAX(impressions) AS Max_Reach FROM mrkt_performance GROUP BY channel")
```

```python
duckdb.sql("SELECT DISTINCT channel, MIN(impressions) AS Min_Reach FROM mrkt_performance GROUP BY channel")
```

**Evaluate Engagement:**

**TOTAL**

```python
duckdb.sql("SELECT DISTINCT(channel),SUM(clicks) AS User_Engagement,SUM(conversions) AS Campaign_Effectiveness FROM mrkt_performance GROUP BY channel")
```

**MAXIMUM**

```python
duckdb.sql("SELECT DISTINCT(channel),MAX(clicks) AS User_Engagement,MAX(conversions) AS Campaign_Effectiveness FROM mrkt_performance GROUP BY channel")
```

**MINIMUM**

```python
duckdb.sql("SELECT DISTINCT(channel),MIN(clicks) AS User_Engagement,MIN(conversions) AS Campaign_Effectiveness FROM mrkt_performance GROUP BY channel")
```

**AVERAGE**

```python
duckdb.sql("SELECT DISTINCT(channel),AVG(clicks) AS User_Engagement,AVG(conversions) AS Campaign_Effectiveness FROM mrkt_performance GROUP BY channel")
```

**Compare Performance of Channels:**

```python
duckdb.sql('''SELECT DISTINCT channel, SUM(impressions) AS No_of_Times_Ads_shown,
SUM(clicks) AS Users_Clicked, AVG(impressions) AS Avg_Times_Shown,
AVG(clicks) AS Avg_Times_Clicked FROM mrkt_performance GROUP BY channel ''')
```

```python
duckdb.sql('''SELECT DISTINCT channel, ROUND(SUM(roas),2) AS Total_Roas,
 ROUND(SUM(revenue_generated),2) AS Total_Revenue, SUM(spend) FROM mrkt_performance
 GROUP BY channel''')
```

```python
duckdb.sql(''' SELECT ROUND(AVG(spend),2) AS Avg_Spend, ROUND(MAX(spend),2) AS Max_Spend, ROUND(MIN(spend),2) AS Min_Spend FROM mrkt_performance ''')
```

```python
duckdb.sql('''SELECT ROUND(AVG(clicks),2) AS Avg_Clicks,MAX(clicks) AS Max_Clicks,MIN(clicks) AS Min_Clicks FROM mrkt_performance ''')
```

```python
duckdb.sql('''SELECT ROUND(AVG(conversions),2) AS AGV_CONVERSIONS,MAX(conversions) AS Max_Conversions ,MIN(conversions)  as Min_Conversions FROM mrkt_performance ''')
```

**Comparing Channels Via clicks and conversions:**

**Below Average**

```python
duckdb.sql('''SELECT campaign_ID, channel, clicks, conversions
FROM mrkt_performance
WHERE clicks <= 550.77 AND conversions <= 55.19''') #between min. and avg.
```

**Above Average**

```python
duckdb.sql('''SELECT campaign_ID, channel, clicks, conversions
FROM mrkt_performance
WHERE clicks >= 550.77 AND conversions >= 55.19''')
```

```python
duckdb.sql('''SELECT DISTINCT(campaign_name),  ROUND(SUM(spend),2) AS SPEND FROM mrkt_performance GROUP BY campaign_name ORDER BY SPEND DESC''')
```
